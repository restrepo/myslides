```python

def cross1(x):
    '''
    x:-> m_{\Sigma^{\pm}}
    cross1 = production cross section \sigma ( p p \to \sigma^{+} \sigma^{-} ) in fb
    '''
    if (100<=x<250):
        a1 = 24.7134440046
        b1 = -3.17385626418
        c1 = -0.0696510981474
        f=exp(a1 + b1*np.log(x) + c1*np.log(x)**2)
    if (250<=x<450):
        a2 = 8.20551597956
        b2 = 3.0480398209
        c2 = -0.656439861563
        f=exp(a2 + b2*np.log(x) + c2*np.log(x)**2)
    if (450<=x<=800):
        a3 = -9.67713364683
        b3 = 9.10638134736
        c3 = -1.16948659419
        f=exp(a3 + b3*np.log(x) + c3*np.log(x)**2)
    if ((0<x<100) or (x>800.0)): 
        f=0.0
    return f 

```
